Null and Alternative Hypotheses (Zomato Cashback Offer)
Null Hypothesis (H₀): The new cashback offer has not increased the average number of daily orders.
H₀: μ ≤ μ₀
Alternative Hypothesis (H₁): The new cashback offer has increased the average number of daily orders.
H₁: μ > μ₀

Where:

μ = Average daily orders after the cashback offer
μ₀ = Average daily orders before the cashback offer

99% Confidence Level (Cricket App)

A 99% confidence level means you want very strong evidence before concluding that match alerts increase user engagement. This corresponds to a 1% significance level (α = 0.01), making it harder to reject the null hypothesis.

Decision Impact:

If the p-value is less than 0.01, reject the null hypothesis and conclude that the match alerts significantly increase user engagement.
If the p-value is greater than or equal to 0.01, fail to reject the null hypothesis because there is not enough evidence at the 99% confidence level to support the feature's effectiveness.

Conclusion: A 99% confidence level reduces the chance of making a false positive (Type I error), but it requires stronger evidence before accepting that the feature improves user engagement.
